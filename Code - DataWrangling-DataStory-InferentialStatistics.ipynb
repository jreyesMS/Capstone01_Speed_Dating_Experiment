## II. Data Wrangling

Initial inspection of the dataframe reveals two types of data:  date-encounter-level data and participant-level data.  The variables from the questionnaire of each speed-date encounter will be isolated and merged with participant demographic data. Follow-up questionaire data will be discarded (columns **attr1_s** thru **amb3_s** ) as more than half of the data is missing.  Will also have to discard the data from followup2/Time3 (columns **you_call - amb5_3**) as more than half of the data is missing.

```python
#Creating the speed-date data frame using demographics and in-date questionnaire data.

#Selecting specific column headings
selections = ['wave','iid','pid','gender','field_cd','age','age_o','race'
              ,'race_o','samerace','goal','date','go_out','match']

#Creating feature data frames from various cross-sections of original dataframe
features01 = data.loc[:, selections] 
features02 = data.loc[:, 'dec':'match_es'] #Participant Rating of parter
features03 = data.loc[:, 'dec_o':'met_o'] #Ratings from participant's partner

#Concatenating the feature dataframes
df = pd.concat([features01, features02, features03], axis=1)

#Inspecting the dataframe
df.info() 
```

```python
#Importing the Imputer to impute the data frame's missing values with the "most frequent values" of each column.
from sklearn.preprocessing import Imputer

#Storing the headings of the dataframe as the imputer converts dataframe into array.
df_headings = df.columns

#Creating and fitting the dataframe to the Imputer
imp = Imputer(missing_values='NaN', strategy='most_frequent', axis=0)
imp.fit(df)

#Creating a new imputed dataframe
df_imp = pd.DataFrame(imp.transform(df))

#Applying the original headings to the dataframe
df_imp.columns = df_headings

#Inspection of the imputed dataframe
df_imp.info()
```

**Observation:**  The **met** and **met_o** have to be corrected as they have values that are outside of the required.  In correcting these values, the an assumption will be made suggesting that both partners will have agreeing responses as to whether or not they have met each other. 

```python
#Correcting the met and met_o columns so that they are 1 = yes and 2 = no

#Correcting the met column
met_incorrect = [3, 5, 6, 7, 8]

df_imp.met.replace(0, 2, inplace=True)
df_imp.met.replace(met_incorrect, 2, inplace=True)


#Correcting the met_o column
met_o_incorrect = [3, 5, 6, 7]

df_imp.met_o.replace(8, 1, inplace=True)
df_imp.met_o.replace(met_o_incorrect, 2, inplace=True)
```
